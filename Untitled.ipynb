```python
from hugr import HugrClient
```

```python
client = HugrClient()
print("URL:", client._url)
print("Auth:", "api_key" if client._api_key else "token" if client._token else "public")
```

```python
client2 = HugrClient.from_connection("local-oidc")
print("Named connection URL:", client2._url)
```

```python
import os
client3 = HugrClient(url=os.environ.get("HUGR_URL", "http://localhost:15004/ipc"))
print("Explicit URL:", client3._url)
```

```python
client4 = HugrClient(
    url="http://localhost:15004/ipc",
    api_key="test-key",
    api_key_header="X-Custom-Api-Key",
)
print("API key header:", client4._api_key_header)
assert client4._api_key_header == "X-Custom-Api-Key"
```

```python
result = client.query("""
{
    core {
        data_sources {
            name
        }
    }
}
""")
```
