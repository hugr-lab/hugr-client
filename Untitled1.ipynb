```python
text_patient_query = '''
query patient_info(
  $patient_id: BigInt = 1, 
  $seen_date: Timestamp = "2025-01-20T00:00:00Z"
  $current_date: Timestamp = "2025-07-20T00:00:00Z"
){
  jq(
    query: """
# ─────────────────────────────── helpers ───────────────────────────────
def age($birth):
  ((now - ($birth)) / 31557600 | floor);

def diag_line:            # объект → "E11.9 – Type-2 diabetes ×7"
  "\(.key.code) – \(.key.description) ×\(.aggregations._rows_count)";

def med_line:             # объект → "Metformin (RX_860975) ×12 – 2025-07-10"
  "\(.key.description) (RX_\(.key.code)) ×\(.aggregations._rows_count) – \(.aggregations.start.min|split("T")[0])";

def vit_line:
  "\(.description): \(.value) \(.units)";

def pref_list($prefix):   # array of objects → "procedure (D_E11.9), procedure2 (D_I10)"
  map((.description // .code) + " (" + $prefix + .code + ")")
  | join(", ") // "";    
# ───────────────────── main transformation ─────────────────────────────
.synthea as $s |
$s.info as $i |
{
  # --- Header ---
  header: ("Patient: "                        +
   (age($i.birthdate_e)|tostring) + "-year-old " +
   ($i.gender|ascii_upcase) + ". "    +
   "Race: " + $i.race + ". "          +
   "Ethnicity: " + $i.ethnicity + ". " +
   "Insurance: " + ($i.payers[0].insurance.name // "Unknown") + ".\n" +
   "Number of encounters: " + ($i.encounters._rows_count|tostring) +
   " (first: " + ($i.encounters.start.min) +
   ", last: "  + ($i.encounters.stop.max) + ")."
  ),
  # --- Diagnoses block ---
  diagnoses: ("⇢ Chronic diagnoses:\n" + ($i.diagnoses // [] | map(diag_line) | join("; "))) ,
  # --- Medications block ---
  medications: ("⇢ Current medications (last 90 d):\n" + ($i.medications // [] | map(med_line) | join("; "))) ,
  # --- Vitals block ---
  vitals: ("Latest vitals: " + ($s.observations // [] | map(vit_line) | join(", "))),
  log:
      ("Encounter log (last " + ($s.encounters|length|tostring) + "):\n" +
       ($s.encounters // []
        | map(
            "• " + (.start) + " - " + (.stop) + " " + (.description // .encounter_class) + " — " +
            (
    			[
        			(if .provider.speciality then " | " + .provider.speciality else "" end),
    				(.conditions // [] | pref_list("D_")),
            		(.procedures // [] | pref_list("P_")),
  					(.medications // [] | pref_list("RX_"))
  				] | map(select(length>0)) | join(" | ")
  			)
          )
        | join("\n") // ""
       ))
} | (.header + "\n\n" + .diagnoses + "\n\n" + .medications + "\n\n" + .vitals + "\n\n" + .log)
    """ 
    include_origin: false) {
    synthea{
      info: patients_by_pk(id: $patient_id){
        payers: ref_payer_transitions_patient_id_id_fkey(
          nested_order_by: [
            {field: "end_date", direction: DESC}
          ]
          nested_limit: 1
        ){
          end_date
          insurance:ref_payer_transitions_payer_id_id_fkey{
            name
          }
        }
        race
        ethnicity
        gender
        birthdate
        birthdate_e: _birthdate_part(extract: epoch)
        geom
        encounters: ref_encounters_patient_id_id_fkey_aggregation{
          start(bucket:day) {
            min
          }
          stop(bucket:day){
            max
          }
          _rows_count
        }
        diagnoses:ref_conditions_patient_id_id_fkey_bucket_aggregation(
          nested_order_by: [
            {field: "aggregations._rows_count", direction: DESC}
          ]
          nested_limit: 5
        ){
          key{
            code
            description
          }
          aggregations{
            _rows_count
          }
        }
        medications: ref_medications_patient_id_id_fkey_bucket_aggregation(
          filter: {
            _or: [
              {start: {gte: $seen_date}}
              {stop: {gte:$current_date}}
            ]
          }
          nested_order_by: [
            {field: "aggregations._rows_count", direction: DESC}
          ]
          nested_limit: 10
        ){
          key{
            code
            description
          }
          aggregations{
            _rows_count
            start{
              min
            }
          }
        }
      }
      observations(
        filter:{
          patient_id: {eq: $patient_id}
          code: {in: [
            "8480-6",  # Systolic BP
            "8462-4",  # Diastolic BP
            "8867-4",  # Heart rate
            "8310-5",  # Body temperature
            "39156-5"  # BMI
          ]}
        }
        distinct_on: ["code"]
        order_by: [
          {field:"code"}
          {field: "date", direction: DESC}
        ]
      ){
        date
        code
        description
        units
        value
      }
      encounters(
        filter: {
          patient_id: {eq: $patient_id}
        }
        order_by:[
          {field: "start", direction: DESC}
        ]
        limit: 8
      ){
        start(bucket: day)
        stop(bucket:day)
        encounter_class
        code
        description
        reason_code
        reason_description
        provider: ref_encounters_provider_id_id_fkey{
          speciality
        }
        conditions: ref_conditions_encounter_id_id_fkey{
          code
          description
        }
        medications: ref_medications_encounter_id_id_fkey{
          code
          description
          dispenses
        }
        procedures: ref_procedures_encounter_id_id_fkey {
          code
          description
        }
      }
    }
  }
}
'''
```

```python
class PatientData:
    def __init__(self, url: str = 'http://localhost:15003/ipc'):
        self.hugr = hugr.HugrClient(url=url)

    def patient_summary(self, id: int) -> str:
        res = self.hugr.query(text_patient_query, {'patient_id': id})
        return res.extension("extensions").content()['jq']['jq']

    async def patients(self):
        conn = hugr.new_stream_connection(self.hugr)
        try:
            async with await conn.stream("{ synthea { patients { id } } }") as stream:
               async for row in stream.rows():
                   yield row['id']
        finally:
            await conn.disconnect()

    async def patient_ids(self):
        ids = []
        async for pid in self.patients():
            ids.append(pid)
        return ids
```

```python
async def collect_patient_ids_async(patient_data):
    ids = []
    async for pid in patient_data.patients():
        ids.append(pid)
    return ids
```

```python
import asyncio, concurrent.futures as cf
from tqdm.asyncio import tqdm_asyncio      # ↔ tqdm>=4.66

MAX_CONCURRENCY = 16       # max parallel requests

async def collect_patient_ids_async(patient_data):
    ids = []
    async for pid in patient_data.patients():
        ids.append(pid)
    return ids

async def gather_patient_texts(patient_data, id_list, max_conc=MAX_CONCURRENCY):
    sem = asyncio.Semaphore(max_conc)
    async def get_one(pid):
        async with sem:
            return pid, await asyncio.to_thread(pd.patient_summary, pid)
    tasks = [asyncio.create_task(get_one(pid)) for pid in id_list]
    texts = {}
    for coro in tqdm_asyncio.as_completed(tasks, total=len(tasks),
                                          desc="pull summaries", unit="patient"):
        pid, txt = await coro
        texts[pid] = txt
    return texts
```
